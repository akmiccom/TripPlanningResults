# データ抽出

```python
import pandas as pd
import csv
```

```python
df = pd.read_csv(f'../csv/searchResults.csv', names=['dep', 'destination', 'date', 'via', 'time', 'price'])
df = df.drop_duplicates(['destination', 'date'])
df = df.dropna()
```

```python
df['price'] = df['price'].str.replace('JP¥|,', '', regex=True).astype(int)
```

```python
display(df[(df['dep'] == 'SYD') & (df['price'] <= 250000)].sort_values(by='price'))
```

```python
for dest in df['destination'].unique():
    display(df[(df['destination'] == dest) & (df['price'] <= 300000)].sort_values(by='price'))
```
