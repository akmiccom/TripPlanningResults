# データ抽出

```python
import pandas as pd
import csv
```

```python
df = pd.read_csv('SYD.csv', names=['destination', 'date', 'via', 'time', 'price'])
df = df.drop_duplicates()
```

```python
df['price'] = df['price'].str.replace('JP¥|,', '', regex=True)
```

```python
for dest in df['destination'].unique():
    display(df[df['destination'] == dest].sort_values(by='price'))
```
